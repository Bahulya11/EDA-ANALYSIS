## Descriptive Statistics

```python
# Mean, median, mode, standard deviation for numerical columns
mean_values = retail.mean(numeric_only=True)
median_values = retail.median(numeric_only=True)
mode_values = retail.mode().iloc[0]
std_dev_values = retail.std(numeric_only=True)
print("Mean Values:\n", mean_values)
print("\nMedian Values:\n", median_values)
print("\nMode Values:\n", mode_values)
print("\nStandard Deviation Values:\n", std_dev_values)
```

## Time Series Analysis

```python
time_series_sales = retail.groupby('Date')['Total Amount'].sum()

# Plotting the sales trend over time
plt.figure(figsize=(10, 6))
time_series_sales.plot(kind='line', title='Sales Trend Over Time', color='blue')
plt.xlabel('Date')
plt.ylabel('Total Sales')
plt.grid()
plt.show()
```

## Customer and Product Analysis

#### Customer Analysis (Gender, Age Group)

```python
# Gender-wise spending
gender_spending = retail.groupby('Gender')['Total Amount'].sum()

# Age group analysis
age_bins = [18, 25, 35, 45, 60, 100]
age_labels = ['18-24', '25-34', '35-44', '45-59', '60+']
retail['Age Group'] = pd.cut(retail['Age'], bins=age_bins, labels=age_labels)
age_group_spending = retail.groupby('Age Group')['Total Amount'].sum()

# Printing insights
print("Gender-wise Spending:\n", gender_spending)
print("\nAge Group Spending:\n", age_group_spending)
```

#### Product Analysis (Category)

```python
# Spending by product category
product_category_spending = retail.groupby('Product Category')['Total Amount'].sum()

# Printing spending by category
print("\nProduct Category Spending:\n", product_category_spending)
```

## Visualization

#### Bar Charts (Gender-wise, Age Group-wise, Product Category-wise)

```python
# Gender-wise spending bar chart
gender_spending.plot(kind='bar', color=['skyblue', 'pink'], title='Gender-wise Spending')
plt.ylabel('Total Sales')
plt.show()

# Age group spending bar chart
age_group_spending.plot(kind='bar', color='orange', title='Age Group-wise Spending')
plt.ylabel('Total Sales')
plt.show()

# Product category spending bar chart
product_category_spending.plot(kind='bar', color='green', title='Product Category Spending')
plt.ylabel('Total Sales')
plt.xticks(rotation=45)
plt.show()
```

#### Heatmap for Sales by Product Category and Gender

```python
# Creating a pivot table for heatmap visualization
pivot_table = retail.pivot_table(
    index='Product Category',
    columns='Gender',
    values='Total Amount',
    aggfunc='sum'
)

# Displaying heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(pivot_table, annot=True, fmt=".0f", cmap="YlGnBu")
plt.title('Total Sales by Product Category and Gender')
plt.show()
```
